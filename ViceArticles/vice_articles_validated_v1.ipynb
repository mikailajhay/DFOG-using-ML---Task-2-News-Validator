### ***Dropping articles whose word count is < 50***

```python
import csv
from collections import Counter
```

```python
column_index = 5
word_count = Counter()
```

```python
with open('vice_articles_10k.csv', 'r') as file:
    reader = csv.reader(file)
    for column in reader:
        words = column[column_index].split()
        word_count.update(words)
```

```python
print(word_count)
```

```python
content = df[df['word count'] < 50]
```

```python
print(content)
```

```python
df.drop(df[df['word count'] < 50].index, inplace=True)
```
