### ***Importing Python libraries***

### ***Checking and dropping of duplicate values***

```python
df[df['title'].duplicated()]['title'].value_counts()
```

```python
df = df.drop_duplicates(subset=['title'], keep='first').reset_index(drop=True)
```

```python
df[df['title'].duplicated()]['title'].value_counts()
```

```python
df.duplicated().sum()
```

```python
num_duplicates = df.duplicated().sum()
```

```python
print("Number of duplicate rows: ", num_duplicates)
```

```python
df.describe(include=['O'])
```

### ***Checking and dropping of missing values***

```python
df.isna().sum()
```

```python
df = df.dropna(subset = ['author', 'article_content', 'content_tags', 'content_tags_link']).reset_index(drop=True)
```

```python
df.isna().sum()
```

```python
df.describe(include=['O'])
```

### ***Checking the number of words per article***

```python
df['word count'] = df['article_content'].str.split().str.len()
```

```python
df.head()
```

### ***Dropping articles whose word count is < 100***

```python
article_content = df[df['word count'] < 100]
```

```python
print(article_content)
```

```python
df.drop(df[df['word count'] < 100].index, inplace=True)
```

```python
df.describe(include=['O'])
```

### ***Exporting the CSV file***

```python
df.to_csv('un_articles_validated_v1.csv', index=False)
```
